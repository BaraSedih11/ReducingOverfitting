```python
import pandas as pd
# Import libraries
from sklearn.linear_model import LinearRegression
from sklearn.datasets import fetch_california_housing
import numpy as np
import matplotlib.pyplot
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv('heart_failure_clinical_records-sample.csv')
df.head()
```

```python
# Define the features and target variable
X = df.drop('DEATH_EVENT', axis=1)  
y = df['DEATH_EVENT']  
```

```python
train_accuracies = []
test_accuracies = []
```

```python

def train_and_evaluate_model(X, y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
    
    model = DecisionTreeClassifier()
    model.fit(X_train, y_train)
    
    y_train_pred = model.predict(X_train)
    train_accuracy = accuracy_score(y_train, y_train_pred)

    y_test_pred = model.predict(X_test)
    test_accuracy = accuracy_score(y_test, y_test_pred)
    
    return train_accuracy, test_accuracy
```

```python
for i in range(7):
    train_accuracy, test_accuracy = train_and_evaluate_model(X, y)
    train_accuracies.append(train_accuracy)
    test_accuracies.append(test_accuracy)
    print('Test' , i, ' : ' , train_accuracy , ' , ' , test_accuracy)
```

```python
param_grid = {
    'max_depth': [3, 5, 7, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf' : [1, 2, 4]
}
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)


model2 = GridSearchCV(estimator=DecisionTreeClassifier(), param_grid=param_grid)
model2.fit(X_train, y_train)
```

```python
best_params = model2.best_params_

print(best_params)


for i in range(7):
    model3 = DecisionTreeClassifier(**best_params)
    model3.fit(X_train, y_train)

    model3Predic = model3.predict(X_train)
    model3Test = model3.predict(X_test)

    model3_train_accuracy = accuracy_score(y_train, model3Predic)
    model3_test_accuracy = accuracy_score(y_test, model3Test)

    print('fianl train acc', model3_train_accuracy)
    print('fianl test acc', model3_test_accuracy)
```
